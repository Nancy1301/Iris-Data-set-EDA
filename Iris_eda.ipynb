# IRIS Dataset EDA
description of dataset from https://www.kaggle.com/arshid/iris-flower-dataset gives the idea of multivariate data set. It contains a set of 150 records under 5 attributes - Petal Length, Petal Width, Sepal Length, Sepal width and Class(Species). The dependent variable would be the species. The independent variable would be: Petal Length, Petal Width, Sepal Length and Sepal width.

Importing all important libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the data into the code

```
data = pd.read_csv("IRIS.csv")
```

# **2-D Scatter plot**
Understanding the scatter graph for each specie.

```
sns.set_style("whitegrid")
sns.FacetGrid(data,hue="species", size =6) \
.map(plt.scatter,"sepal_length","sepal_width")\
.add_legend()
plt.show()
```

we can observe that if the sepal width is in between 3-4.5 and length is in between 4.5-6, Iris-setosa would be the species.
the rest two species are overlapping, hence no conclusion.

## Pair Plot

```
sns.set_style("whitegrid")
sns.pairplot(data, hue="species", size =3)
plt.show();
```

if petal length is less than equal to 2 and width is less than equal to 1, the species would be setosa.
similarly, if petal width is less than 2 and greater than 1 and its length is lying between 2.5 and 5. the species would be versicolor.

Univariate Analysis

CONCLUSION would state that: Analysing petal length is much better than petal width which is in turn more efficient than sepal length and width.

FINDING MEANS:

```
iris_setosa = data.loc[data["species"] == "Iris-setosa"]
iris_virginica = data.loc[data["species"] == "Iris-virginica"]
iris_versicolor = data.loc[data["species"] == "Iris-versicolor"]
```

```
#Mean, Variance, Std-deviation,  
print("Means:")
print(np.mean(iris_setosa["petal_length"]))
#Mean with an outlier.
print(np.mean(np.append(iris_setosa["petal_length"],50)));
print(np.mean(iris_virginica["petal_length"]))
print(np.mean(iris_versicolor["petal_length"]))
```

We can conclude that the petal length of iris setosa has its mean less as compared to iris virginica and versicolor.

# Median, Percentile, Quantile, IQR, MAD

```
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(iris_setosa["petal_length"]))
#Median with an outlier
print(np.median(np.append(iris_setosa["petal_length"],50)));
print(np.median(iris_virginica["petal_length"]))
print(np.median(iris_versicolor["petal_length"]))


print("\nQuantiles:")
print(np.percentile(iris_setosa["petal_length"],np.arange(0, 100, 25)))
print(np.percentile(iris_virginica["petal_length"],np.arange(0, 100, 25)))
print(np.percentile(iris_versicolor["petal_length"], np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(iris_setosa["petal_length"],90))
print(np.percentile(iris_virginica["petal_length"],90))
print(np.percentile(iris_versicolor["petal_length"], 90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(iris_setosa["petal_length"]))
print(robust.mad(iris_virginica["petal_length"]))
print(robust.mad(iris_versicolor["petal_length"]))
```
